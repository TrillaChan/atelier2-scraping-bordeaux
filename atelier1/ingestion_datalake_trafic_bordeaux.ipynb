```python
from pathlib import Path

ics_path = Path.home() / "Downloads" / "Zone-A.ics"
ics_path, ics_path.exists()
```

```python
with open(ics_path, "r", encoding="utf-8") as f:
    lines = f.readlines()

lines[:20]
```

```python
events = []
current = {}

for line in lines:
    line = line.strip()
    
    if line == "BEGIN:VEVENT":
        current = {}
    elif line.startswith("SUMMARY:"):
        current["event"] = line.replace("SUMMARY:", "")
    elif line.startswith("DTSTART"):
        current["start"] = line.split(":")[1]
    elif line.startswith("DTEND"):
        current["end"] = line.split(":")[1]
    elif line == "END:VEVENT":
        events.append(current)

import pandas as pd
df_vacances = pd.DataFrame(events)

df_vacances.head()
```

```python
from datetime import datetime

output_dir = Path.cwd() / "datalake" / "parquet"
output_dir.mkdir(parents=True, exist_ok=True)

run_date = datetime.now().strftime("%Y%m%d")

parquet_path = output_dir / f"calendrier_vacances_scolaires_zoneA_{run_date}.parquet"
df_vacances.to_parquet(parquet_path, index=False)

parquet_path
```
